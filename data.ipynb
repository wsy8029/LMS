# 데이터 생성



#### UserID: 사용자 ID / 210001, 210002, 210003 ...
#### Sco: 학습 성취도(종합 점수) / 0~100 (점)
#### Con: 소요시간(집중도) / 0~100 (분)
#### Pas: TC통과율(통과 유무) 0~100 (%)


```python
import pandas as pd
import numpy as np
```

```python
# UserID: 사용자 ID / 210001, 210002, 210003 ...
# Sco: 학습 성취도(종합 점수) / 0~100 (점)
# Con: 소요시간(집중도) / 0~100 (분)
# Pas: TC통과율(통과 유무) 0~100 (%)
feature = ['UserID', 
           'Sco001', 'Con001', 'Pas001', 
           'Sco002', 'Con002', 'Pas002', 
           'Sco003', 'Con003', 'Pas003', 
           'Sco004', 'Con004', 'Pas004', 
           'Sco005', 'Con005', 'Pas005', 
           'Sco006', 'Con006', 'Pas006', 
           'Sco007', 'Con007', 'Pas007', 
           'Sco008', 'Con008', 'Pas008', 
           'Sco009', 'Con009', 'Pas009', 
           'Sco010', 'Con010', 'Pas010']

users = 330
```

```python
df = pd.DataFrame(data=np.random.randint(0,100,(users, len(feature))),
                 columns=feature)
```

- 랜덤으로 생성할 경우, 너무 연관성 없이 데이터가 랜덤하게 생성되기 때문에, 경향을 추가해서 생성하기 위한 코드
- 총 330명 데이터 생성
- 학생 1명씩 순서대로 생성 
- User ID는 반복문 내에서 순서대로 생성 
- 점수 분포는, 상 99명(30%), 중 132명(40%), 하 99명(30%) 분포로 생성

#### 상 99명: Sco 범위 70-100, Con 범위 10-40, Pas 범위 80-100. Sco = 0.4*(100-Con) + 0.6*Pas 실제  72-96
#### 중 132명: Sco 범위 40-70, Con 범위 30-60, Pas 범위 50-80. Sco = 0.4*(100-con) + 0.6*Pas 실제 46-72
#### 하 99명: Sco 범위 0-40, Con 범위 50-100, Pas 범위 0-50. Sco = 0.4*(100-con) + 0.6*Pas 실제 0-50

```python
# 랜덤으로 생성할 경우, 너무 연관성 없이 데이터가 랜덤하게 생성되기 때문에, 경향을 추가해서 생성하기 위한 코드
# 총 330명 데이터 생성
# 학생 1명씩 순서대로 생성 
# User ID는 반복문 내에서 순서대로 생성 
# 점수 분포는, 상 99명(30%), 중 132명(40%), 하 99명(30%) 분포로 생성
# 상 99명: Sco 범위 70~100, Con 범위 10~40, Pas 범위 80~100. Sco = 0.4*(100-Con) + 0.6*Pas 실제  72~96
# 중 132명: Sco 범위 40~70, Con 범위 30~60, Pas 범위 50~80. Sco = 0.4*(100-con) + 0.6*Pas 실제 46~72
# 하 99명: Sco 범위 0~40, Con 범위 50~100, Pas 범위 0~50. Sco = 0.4*(100-con) + 0.6*Pas 실제 0~50
```

```python
# 210001~210330 사이에서 중복 없는 난수 리스트 생성
import random

userid = []
rand_num = random.randint(210001,210330)

for i in range(users):
    while rand_num in userid:
        rand_num = random.randint(210001,210330)
    userid.append(rand_num)
    
```

```python
df = pd.DataFrame(columns=feature)

high = int(users*0.3)
mid = int(users*0.4)
low = int(users*0.3)


# 상 등급 학생 데이터 생성
for i in range(high):
    user_data = []
    user_data.append(userid[i])
    score_list = []
    for i in range(10):
        con = random.randint(10,40)
        pas = random.randint(80,100)
        sco = int(0.4*(100-con)+0.6*pas)
        score_list.extend([sco,con,pas])
    user_data.extend(score_list)
    df.loc[len(df)] = user_data

# 중 등급 학생 데이터 생성
for i in range(mid):
    user_data = []
    user_data.append(userid[i])
    score_list = []
    for i in range(10):
        con = random.randint(30,60)
        pas = random.randint(50,80)
        sco = int(0.4*(100-con)+0.6*pas)
        score_list.extend([sco,con,pas])
    user_data.extend(score_list)
    df.loc[len(df)] = user_data
    
# 하 등급 학생 데이터 생성
for i in range(low):
    user_data = []
    user_data.append(userid[i])
    score_list = []
    for i in range(10):
        con = random.randint(50,100)
        pas = random.randint(0,50)
        sco = int(0.4*(100-con)+0.6*pas)
        score_list.extend([sco,con,pas])
    user_data.extend(score_list)
    df.loc[len(df)] = user_data
    
    
```
